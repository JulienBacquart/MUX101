```python
import os

WIDTH = 600
HEIGHT = 600

NUM_TILES = 10
tile_size = WIDTH / NUM_TILES

def setup():
    size(WIDTH, HEIGHT)

    no_stroke()
    no_loop()

def draw():
    for x in range(0, NUM_TILES):
        for y in range(0, NUM_TILES):
            position = Py5Vector2D(x * tile_size, y * tile_size)

            # Pick a random shade of grey for the rectangle
            col = random_int(255)
            fill(col)

            push_matrix()
            translate(position.x, position.y)
            rect(0, 0, tile_size, tile_size)

            # Draw the cental circle in the 'opposite' shade of grey
            fill(255-col)
            ellipse(tile_size/2, tile_size/2, 0.9*tile_size, 0.9*tile_size)
            pop_matrix()

    save_frame(os.getcwd() + "/random_squares_####.png", use_thread=True)

run_sketch()
```
