```python
import os

WIDTH = 600
NUM_TILES = 8

tile_size = WIDTH / NUM_TILES
var_stroke_weight = tile_size * 0.39

NUM_BLACK_BAND = 4
    
def setup():
    size(WIDTH, WIDTH)
    smooth()
    
    background(0)
    stroke(255)
    stroke_weight(var_stroke_weight)
#   stroke_cap(ROUND)
    stroke_cap(SQUARE)
    
    no_fill()
    no_loop()

def draw():
    num_band = 2 * NUM_BLACK_BAND + 1
    
    for x in range(0, NUM_TILES):
        for y in range(0, NUM_TILES):
            # start with white
            col = 255
                        
            with push_matrix():
                translate(x * tile_size + tile_size/2, y * tile_size + tile_size/2)
                rotate(random_int() * HALF_PI)
                
                # We draw a white arc, then a smaller black arc inside, then a smaller one 
                # and so on ...
                for i in range(num_band, 0, -2):                
                    stroke(col) 
                    stroke_weight((i / num_band) * var_stroke_weight)
                
                    arc(-tile_size/2, -tile_size/2, tile_size, tile_size, 0, HALF_PI)
                    arc(tile_size/2, tile_size/2, tile_size, tile_size, PI, 3 * HALF_PI) 

                    # flip the color
                    col = 255 - col
            
    save_frame(os.getcwd() + "/truchet_tiles_####.png", use_thread=True)

run_sketch()
```
