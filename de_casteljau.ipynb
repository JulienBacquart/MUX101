```python
from collections import namedtuple

WIDTH = 600
HEIGHT = 600

D = 2

Point = namedtuple('Point', ['x', 'y'])

p0 = Point(100, 450)
p1 = Point(150, 250)
p2 = Point(450, 250)
p3 = Point(500, 450)

def middle_point(p0, p1):
    return Point((p0.x + p1.x)/2, (p0.y + p1.y)/2)

def de_casteljau(p0, p1, p2, p3):
    if (dist(*p0, *p1) >= D):  
        stroke(255, 100, 0)
        line(*p0, *p1)
        line(*p1, *p2)
        line(*p2, *p3)

        l1 = middle_point(p0, p1)
        r1 = middle_point(p2, p3)
        h = middle_point(p1, p2)

        stroke(100, 255, 0);
        line(*l1, *h)
        line(*r1, *h)

        l2 = middle_point(l1, h)
        r2 = middle_point(r1, h)

        l3 = middle_point(l2, r2)
        
        de_casteljau(p0, l1, l2, l3)
        de_casteljau(p3, r1, r2, l3)
    else:
        stroke(255, 255, 255);
        line(*p0, *p1)
        line(*p1, *p2)
        line(*p2, *p3)
        

def setup():
    size(WIDTH, HEIGHT)
    no_fill()
    no_loop()
    
def draw():
    background(0)
    p1 = Point(mouse_x, mouse_y)
 
    de_casteljau(p0, p1, p2, p3)

def mouse_moved():
    redraw()
    
run_sketch()
```
