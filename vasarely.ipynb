```python
import os

WIDTH = 600
HEIGHT = 600
TILE_SIZE = 50

num_tile_width = WIDTH // TILE_SIZE
half_diag = sqrt(pow(TILE_SIZE//2, 2) * 2) # Calculate the hypotenuse
num_tile_height = ceil(HEIGHT / half_diag) + 1

palette = ['#5e3b7f', '#b55ea7', '#ebaf04', '#2a43a6', '#765b2b', '#8a2c3a', '#2b4239', '#6a3e99', '#c462b0', '#bd8e0b', '#4659e2', '#d39504', '#34a414', '#aabc11', '#a12e36', '#3fa704', '#e3e208', '#319c0b', '#2e34da', '#d0546d', '#dc440a', '#cc3c1f', '#4667c6', '#2c5974', '#9c549c', '#1d6b3c', '#7966b5', '#f55404', '#af467a', '#7a3d5a', '#b44f1c', '#2a4c3a', '#52673a', '#bdbdbb', '#2c14cb', '#af9460', '#3d962a', '#be342a', '#853d76', '#948cc4', '#3020b8', '#a8344f', '#5234c5', '#a4542c', '#dd5d05', '#2c2c8c', '#1f7c35', '#dc4434', '#208434', '#748c14', '#1c7c2c', '#2c24d6', '#ec4c04', '#3c34d0', '#246c54', '#dcdc64', '#7c6414', '#3c7436', '#a43c2c', '#2c3cc4', '#7c746c', '#245c3c', '#6c343c']

def setup():
    size(WIDTH, HEIGHT)
    
    rect_mode(CENTER)

    no_stroke()
    no_loop()

def draw():    
    for x in range(0, num_tile_width):
        for y in range(0, num_tile_height):
            # even or odd vertical line ?
            x_offset = half_diag if ((y % 2) == 1) else 0
            
            # if odd vertical line, we translate by half a square 
            position = Py5Vector2D(x * half_diag * 2 + x_offset, y * half_diag)

            with push_matrix():
                # Pick a random color in the palette
                col = random_choice(palette)
                fill(col)

                # Draw the rectangle
                translate(position.x, position.y)
                rotate(PI/4) # rotate 45 degrees
                rect(0, 0, TILE_SIZE, TILE_SIZE)

                # Pick a random color in the palette
                col = random_choice(palette)
                fill(col)

                # Draw the cental circle
                fill(col)            
                circle(0, 0, random(0.7, 1.1) * half_diag)

    save_frame(os.getcwd() + "/vasarely_####.png", use_thread=True)

run_sketch()
```
