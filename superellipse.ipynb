```python
import numpy as np
import py5_tools

WIDTH = 300
HEIGHT = 300

a = WIDTH // 2
b = HEIGHT // 2

num_point = 500
num_n = 50

n_values = np.linspace(0.75, 1.5, num=num_n)
n_indice = 0
n_sum = 0

def setup():
    size(WIDTH, HEIGHT)
    background(0)
    stroke(255)
    
def draw():
    global n_indice, n_sum
    
    translate(width/2, height/2)

    if n_indice < len(n_values):
        n = n_values[n_indice]
        n_sum += n
        n_indice += 1

        rotate(n_sum * PI/num_n)
        scale(remap(n, 0.75, 1.5, 0.1, 1.5))
        
        for t in np.linspace(0, HALF_PI/2, num=num_point):
            x = a * abs(cos(t)) ** (2/n)
            y = b * abs(sin(t)) ** (2/n)
            
            stroke(remap(x, a, a/sqrt(2), 0, 255))
            point(x, y)
            point(y, x)
            point(-x, y)
            point(y, -x)
            point(x, -y)
            point(-y, x)
            point(-x, -y)
            point(-y, -x)
        
    else:
        no_loop()
    
run_sketch()

py5_tools.animated_gif('./superellipse.gif', count= num_n, period=0.01, duration=.04)
```
