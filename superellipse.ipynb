```python
import numpy as np
import py5_tools

WIDTH = 400
HEIGHT = 400

a = WIDTH // 2
b = HEIGHT // 2

num_point = 500
num_n = 45

n_values = np.linspace(0.75, 2, num=num_n)
n_indice = 0
n_sum = 0

def setup():
    size(WIDTH, HEIGHT)
    background(0)
    stroke(255)
    
def draw():
    global n_indice, n_sum
    
    translate(width/2, height/2)

    if n_indice < len(n_values):
        n = n_values[n_indice]
        n_sum += n
        n_indice += 1

        rotate(n_sum * PI/num_n)
        
        for t in np.linspace(0, TAU, num=num_point):
            x = a * abs(cos(t)) ** (2/n)
            y = b * abs(sin(t)) ** (2/n)

            stroke(remap(x, 0, a, 0, 255))
            point(x, y)
            point(-x, y)
            point(x, -y)
            point(-x, -y)
        
    else:
        no_loop()
    
run_sketch()

py5_tools.animated_gif('./superellipse.gif', count= num_n, period=0.01, duration=.04)
```
